```python
import os
```

```python
# Get the current working directory (should be the folder you're in)
folder_path = os.getcwd()
```

```python
#add +1bond +2angles 4th and 5th lines
# Iterate over all files in the current directory
for filename in os.listdir(folder_path):
    # Process only the '_post.txt' files
    if filename.endswith('_post.txt'):
        # Construct the full path to the file
        file_path = os.path.join(folder_path, filename)
        
        with open(file_path, 'r') as file:
            # Read all lines in the file
            lines = file.readlines()

        # Modify the lines corresponding to the number of bonds and angles
        lines[3] = f"     {int(lines[3].split()[0]) + 1} bonds\n"
        lines[4] = f"     {int(lines[4].split()[0]) + 2} angles\n"

        # Write the modified lines back to the file
        with open(file_path, 'w') as file:
            file.writelines(lines)
        
        print(f'Updated: {filename}')
```

```python
#add new bond line X 3 4 3
# Iterate over all files in the current directory
for filename in os.listdir(folder_path):
    # Process only the '_post.txt' files
    if filename.endswith('_post.txt'):
        # Construct the full path to the file
        file_path = os.path.join(folder_path, filename)
        
        with open(file_path, 'r') as file:
            # Read all lines in the file
            lines = file.readlines()

        # Find where the Bonds section ends by searching for the line after "Angles"
        bond_end_line = None
        for i, line in enumerate(lines):
            if line.startswith("Angles"):
                bond_end_line = i
                break

        # Insert the new bond line after the last bond line
        if bond_end_line:
            bond_count = int(lines[3].split()[0])  # Get the bond count from the 4th line
            new_bond_line = f"     {bond_count} 3 4 2\n"  # New bond line to add
            lines.insert(bond_end_line, new_bond_line)  # Insert the new bond line before "Angles"

        # Write the modified lines back to the file
        with open(file_path, 'w') as file:
            file.writelines(lines)
        
        print(f'Updated: {filename}')
```

```python
#add two angles
# Loop through all _post.txt files
for filename in os.listdir(folder_path):
    if filename.endswith('_post.txt'):
        file_path = os.path.join(folder_path, filename)
        
        with open(file_path, 'r') as file:
            lines = file.readlines()

        # Get updated number of angles from line 5
        angle_count = int(lines[4].split()[0])

        # Find where the Angles section starts
        angle_section_start = None
        for i, line in enumerate(lines):
            if line.strip() == "Angles":
                angle_section_start = i
                break

        if angle_section_start is not None:
            # Angles data starts two lines after the "Angles" header
            angle_data_start = angle_section_start + 2

            # Find end of Angles section by scanning until a blank line or end of file
            angle_data_end = angle_data_start
            while angle_data_end < len(lines) and lines[angle_data_end].strip():
                angle_data_end += 1

            # Add the two new angles
            new_angle_1 = f"     {angle_count} 2 5 4 2\n"
            new_angle_2 = f"     {angle_count + 1} 2 6 2 4\n"
            lines.insert(angle_data_end, new_angle_1)
            lines.insert(angle_data_end + 1, new_angle_2)

            # Write modified content back to the file
            with open(file_path, 'w') as file:
                file.writelines(lines)

            print(f"Updated angles in: {filename}")
```

```python
for filename in os.listdir(folder_path):
    if filename.endswith('_post.txt'):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, 'r') as file:
            lines = file.readlines()

        # Modify lines 9, 10, 11, 12 (Python indexing: 8 to 11)
        for i in range(8, 12):
            parts = lines[i].split()
            if len(parts) >= 2 and parts[1] == '7':
                parts[1] = '8'
                lines[i] = '    ' + ' '.join(parts) + '\n'

        with open(file_path, 'w') as file:
            file.writelines(lines)

        print(f"Updated types in lines 9–12 of: {filename}")
```
